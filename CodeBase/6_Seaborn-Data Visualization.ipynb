### Distribution plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')#loaded data sets
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])#used for single measure(bar chart)
```

```python
#histogram
sns.distplot(tips['total_bill'],kde=False)
```

```python
tips.shape
```

```python
#adding bin to histogram
sns.distplot(tips['total_bill'],kde=False,bins=50)
```

```python
#Joint plot
#combine two distribution(two measure)
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#affect inside-what is happening , where is the crowd
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')#default is scatter
```

```python
#adding regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#showing density
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#pair wise relationship of every possible(,easures only-numeric)
#pairplot
sns.pairplot(tips)
```

```python
#adding category
sns.pairplot(tips,hue='sex')
```

```python
#rugplot
#show color denseity that reflects higher or lower value-similar to displot/bar chart  which show exact value of y axis
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

#### kde--   Kernel Density estimation

```python
sns.kdeplot(tips['total_bill'],)
#sns.color_palette("Red")
```

```python
sns.palplot(sns.color_palette("Paired"))#?
```

```python
sns.kdeplot(tips['total_bill'],tips['tip'])
```

## Categorical Plots

```python
#Dimensions/categroies
tips.head()
```

```python
#sex,gender,smoker,day,time
```

```python
#1 bar plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#adding estimate,avg
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#count plot- count the numberr of occurence
sns.countplot(x='sex',data=tips)
```

```python
#box plot and viloin plot-categorical data
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#hue-add more category
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#violin-allows us to plt all the components of the data point
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#strip plot-scatter plot where one variable is numerical
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#add jitter- add noise- seperate values
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
#swar plot-scatter+violin
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
#factor plot- General plottong method
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

## Grids

```python
#iris datase
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris) ##Pairgrid gives us control over Pairplot(types of plot we want to plot)
```

```python
import matplotlib.pyplot as plt
g= sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
