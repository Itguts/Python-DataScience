# Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='XY')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='X cube')
ax.legend()
```

```python
#upper right
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=(0.1,0.1))
```

# Plot Appearance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='orange')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00') #RGB Hex colorcode
```

```python
#linwidth
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',linewidth=3) 
```

```python
#transparent-alpha
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',lw=3,alpha=0.5) 
```

```python
#line style-dash line, dot line
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',linewidth=3,linestyle=":") 
```

```python
#or
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',lw=3,ls=":") 
```

```python
#markup
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',linewidth=3,linestyle=":",marker='o')#each point of x,y 
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',linewidth=3,linestyle=":",marker='o',markersize=12)#each point of x,y 
```

```python
#marker face color,edge,edge color
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',linewidth=3,linestyle=":",marker='o',markersize=12,
          markerfacecolor='green',markeredgewidth=3,markeredgecolor='red')#each point of x,y 
```

# Plot range- control over axis appearnace

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF8C00',lw=3,ls="-") 
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
